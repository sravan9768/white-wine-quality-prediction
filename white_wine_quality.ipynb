# Preprocessing

```python
# Columns to handle outliers in
columns_to_handle = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
                    'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
                    'pH', 'sulphates']

# Function to replace outliers with median
def replace_outliers_with_median(data, column):
    """Replaces outliers in a column with the median value.

    Args:
        data: DataFrame
        column: Name of the column

    Returns:
        DataFrame with outliers replaced by median
    """
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    median_value = data[column].median()
    data.loc[(data[column] < lower_bound) | (data[column] > upper_bound), column] = median_value

    return data

# Replace outliers in the specified columns
for column in columns_to_handle:
    df = replace_outliers_with_median(df, column)
```

```python
def winsorize_column(data, column, lower_percentile=1, upper_percentile=99):
    """Winsorizes a column in a DataFrame.

    Args:
        data: DataFrame
        column: Name of the column to winsorize
        lower_percentile: Lower percentile for capping
        upper_percentile: Upper percentile for capping

    Returns:
        DataFrame with the winsorized column
    """
    lower_bound = data[column].quantile(lower_percentile / 100)
    upper_bound = data[column].quantile(upper_percentile / 100)
    data[column] = np.clip(data[column], lower_bound, upper_bound)
    return data

# Columns to apply capping to
columns_to_cap = ['volatile acidity', 'citric acid', 'chlorides', 'free sulfur dioxide', 'pH', 'sulphates']

# Apply capping to the specified columns
for column in columns_to_cap:
    df = winsorize_column(df, column)
```

```python


# Apply log transformation to 'citric acid' and 'sulphates'
df['citric acid'] = np.log1p(df['citric acid'])  # Use log1p to handle zero values
df['sulphates'] = np.log1p(df['sulphates'])
```

```python
def remove_outliers_iqr(data, column):
    """Removes outliers from a column using the IQR method.

    Args:
        data: DataFrame
        column: Name of the column

    Returns:
        DataFrame with outliers removed
    """
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]

    return data

# Remove outliers from 'citric acid'
df = remove_outliers_iqr(df, 'citric acid')
df = remove_outliers_iqr(df, 'residual sugar')

```

# Define target variable(y) and feature variable(x)

```python
x=df.drop(columns='quality',axis=1)
y=df['quality']
```

```python
print(x.shape)
print(y.shape)
```

# Train Test split

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Modelling

```python
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression(penalty='l2', C=1.0, solver='liblinear')
lg.fit(x_train, y_train)

```

# Get model prediction

```python
y_pred_train = lg.predict(x_train)  # Predictions on training data
y_pred_test = lg.predict(x_test)    # Predictions on testing data
y_pred=lg.predict(x_test)
```

# Model evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
#  Training Accuracy
train_accuracy = accuracy_score(y_train, y_pred_train)
print(f"Training Accuracy: {train_accuracy}")

# Testing Accuracy
test_accuracy = accuracy_score(y_test, y_pred_test)
print(f"Testing Accuracy: {test_accuracy}")

# Overall Accuracy (on the entire dataset)
overall_accuracy = accuracy_score(y, lg.predict(x))
print(f"Overall Accuracy: {overall_accuracy}")

#  Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix:\n", conf_matrix)

#  Classification Report
class_report = classification_report(y_test, y_pred_test)
print("Classification Report:\n", class_report)
```

```python

from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=5)
rf_classifier.fit(x_train, y_train)

y_pred_train = rf_classifier.predict(x_train)
y_pred_test = rf_classifier.predict(x_test)

train_accuracy = accuracy_score(y_train, y_pred_train)
print(f"Training Accuracy: {train_accuracy}")

test_accuracy = accuracy_score(y_test, y_pred_test)
print(f"Testing Accuracy: {test_accuracy}")
overall_accuracy = accuracy_score(y, rf_classifier.predict(x))
print(f"Overall Accuracy: {overall_accuracy}")

conf_matrix = confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix:\n", conf_matrix)

class_report = classification_report(y_test, y_pred_test)
print("Classification Report:\n", class_report)
```

## ROC AOC curve


```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Assuming 'rf_classifier' is your trained Random Forest model,
# 'x_test' is your test data, and 'y_test' are the true labels

# Get predicted probabilities for both classes
y_probs = rf_classifier.predict_proba(x_test)

# Calculate ROC curve values for class 0
fpr_0, tpr_0, thresholds_0 = roc_curve(y_test, y_probs[:, 0], pos_label=0)
roc_auc_0 = roc_auc_score(y_test, y_probs[:, 0])

# Calculate ROC curve values for class 1
fpr_1, tpr_1, thresholds_1 = roc_curve(y_test, y_probs[:, 1], pos_label=1)
roc_auc_1 = roc_auc_score(y_test, y_probs[:, 1])

# Plot the ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_0, tpr_0, color='darkorange', lw=2, label=f'ROC curve (Class 0, AUC = {roc_auc_0:.2f})')
plt.plot(fpr_1, tpr_1, color='darkblue', lw=2, label=f'ROC curve (Class 1, AUC = {roc_auc_1:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line (random classifier)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
